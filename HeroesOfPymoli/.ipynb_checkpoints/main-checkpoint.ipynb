```python
import pandas as pd
import csv
```

```python
data = r"purchase_data.csv"
```

```python
purchasedata_df = pd.read_csv(data)
purchasedata_df.head()
```

```python
#Player Count
Total_Players_dict = [{"Total Players" : (len(set(purchasedata_df["SN"])))}]
total_Players = pd.DataFrame(Total_Players_dict)
total_Players
```

```python
#Purchasing Analysis (Total)
summary_dict = [{"Number of Unique items" : (purchasedata_df["Item Name"].nunique()),
               "Average Price" :(purchasedata_df["Price"].mean()),
              "Number of purchases": (purchasedata_df["Purchase ID"].nunique()),
              "Total Revenue" :(purchasedata_df["Price"].sum()),
              }
             ]
summary_df = pd.DataFrame(summary_dict)
summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:.2f}".format)
summary_df
```

```python
#Gender Demographics
# Grouping purchase data by Gender. This is an object
gendercount = purchasedata_df.groupby("Gender")
# Count the total of screen names "SN" by gender. This is a number
totalgendercount = gendercount.nunique()["SN"] 
# Total gender count divided by total players 
percentage_of_players = totalgendercount / len(set(purchasedata_df["SN"])) * 100

# Create data frame with obtained values
gender_demographics_df = pd.DataFrame({"Total Count": totalgendercount, "Percentage of Players": percentage_of_players})
gender_demographics_df.index.name = None

# Format the values sorted by total count in descending order, and two decimal places for the percentage
gender_demographics_df.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"{:.2f}%"})
```

```python
# Purchasing Analysis (Gender)
#calculating purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#This is a groupby object. We use count, sum, mean functions on this to find the values of the stuff inside the object which is the genders
gendercount = purchasedata_df.groupby("Gender") 
purchasecount = gendercount["Purchase ID"].count()
AvgPurchaseprice = gendercount["Price"].mean()
TotalpurchaseValue = gendercount["Price"].sum()
AvgPurchasetotalperperson = TotalpurchaseValue / totalgendercount

Gender_Purchasing_Analysis_dict = {"Purchase Count" : purchasecount,
                                    "Average Purchase Price" :AvgPurchaseprice,
                                    "Total Purchase": TotalpurchaseValue,
                                    "Average Purchase Total Per Person" :AvgPurchasetotalperperson
                                  }
Gender_Purchasing_Analysis_df = pd.DataFrame(Gender_Purchasing_Analysis_dict)
Gender_Purchasing_Analysis_df["Average Purchase Price"] = Gender_Purchasing_Analysis_df["Average Purchase Price"].map("${:.2f}".format)
Gender_Purchasing_Analysis_df["Total Purchase"] = Gender_Purchasing_Analysis_df["Total Purchase"].map("${:.2f}".format)
Gender_Purchasing_Analysis_df["Average Purchase Total Per Person"] = Gender_Purchasing_Analysis_df["Average Purchase Total Per Person"].map("${:.2f}".format)
Gender_Purchasing_Analysis_df
```

```python
#Age Demographics
#Calculating the numbers and percentages by age group
age_bins = [0, 9.9, 14.9, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_group = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+" ]
purchasedata_df["Age Group"] = pd.cut(purchasedata_df['Age'], bins=age_bins, labels=age_group)
#This is a dataframe
purchasedata_df
#The groupedby dataframe is an object
purchasedataGroupedbyAge = purchasedata_df.groupby("Age Group")
#number of players in each age group. This will be a list
AgeGroupTotals = purchasedataGroupedbyAge.nunique()["SN"] 
#Percentage of players in each age group. This will be a list
Percentage_of_Players = AgeGroupTotals/len(set(purchasedata_df["SN"])) *100
AgeDemographichs_df = pd.DataFrame({"Total Count": AgeGroupTotals, "Percentage of Players": Percentage_of_Players})
AgeDemographichs_df["Percentage of Players"] = AgeDemographichs_df["Percentage of Players"].map("{:.2f}%".format)

AgeDemographichs_df
```

```python
#Purchasing Analysis (Age)
#calculations to obtain purchase count, avg. purchase price, avg. purchase total per person

purchasedata_df
#The groupedby dataframe is an object
purchasedataGroupedbyAge = purchasedata_df.groupby("Age Group")
#The below are lists
purchaseCntbyAge = purchasedataGroupedbyAge["Purchase ID"].count()
AvgPurchPriceforAgeGroup = purchasedataGroupedbyAge["Price"].mean()
TotalPurchaseValueforAgeGroup = purchasedataGroupedbyAge["Price"].sum()
AveragePurchaseTotalperPersoninAgeGroup = TotalPurchaseValueforAgeGroup/AgeGroupTotals

purchasingAnlysisAge_dict = {"Purchase Count" : purchaseCntbyAge, 
                             "Average Purchase Price" : AvgPurchPriceforAgeGroup, 
                             "Total Purchase Value" : TotalPurchaseValueforAgeGroup, 
                             "Avg Total Purchase per Person":AveragePurchaseTotalperPersoninAgeGroup
                            }

PurchasingAnlysisAge_df = pd.DataFrame(purchasingAnlysisAge_dict)
PurchasingAnlysisAge_df["Total Purchase Value"] = PurchasingAnlysisAge_df["Total Purchase Value"].map("${:.2f}".format)
PurchasingAnlysisAge_df["Average Purchase Price"] = PurchasingAnlysisAge_df["Average Purchase Price"].map("${:.2f}".format)
PurchasingAnlysisAge_df["Avg Total Purchase per Person"] = PurchasingAnlysisAge_df["Avg Total Purchase per Person"].map("${:.2f}".format)
PurchasingAnlysisAge_df
```

```python
#Top Spenders
#Purchase Count, Average Purchase Price, Total Purchase Value
purchasedata_df
#The groupedby dataframe is an object
purchasedataGroupedbySN = purchasedata_df.groupby("SN")

#List showing count of purchase by each screenname
purchasecountbySN = purchasedataGroupedbySN["Purchase ID"].count()
AveragePurchasePriceforSN = purchasedataGroupedbySN["Price"].mean()
TotalPurchaseValueSN = purchasedataGroupedbySN["Price"].sum()

top_spenders_dict = {"Purchase Count":purchasecountbySN,
                     "Average Purchase Price":AveragePurchasePriceforSN, 
                     "Total Purchase Value":TotalPurchaseValueSN
                     }

top_spenders_df = pd.DataFrame(top_spenders_dict)
top_spenders_df_desc = top_spenders_df.sort_values(["Total Purchase Value"], ascending = False)

top_spenders_df_desc["Average Purchase Price"] = top_spenders_df_desc["Average Purchase Price"].map("${:.2f}".format)
top_spenders_df_desc["Total Purchase Value"] = top_spenders_df_desc["Total Purchase Value"].map("${:.2f}".format)

top_spenders_df_desc.head()
```

```python
#Most Popular Items
#Retrieving the Item ID, Item Name, and Item Price columns
purchasedata_Itm_df = purchasedata_df[["Item ID", "Item Name", "Price"]]
purchasedata_Itm_df
#Group by Item ID and Item Name. 
purchasedata_Itm = purchasedata_Itm_df.groupby(["Item ID", "Item Name"])
#Perform calculations to obtain purchase count, average item price, and total purchase value
#These are lists
purchasecnt = purchasedata_Itm["Item ID"].count()
totalPurchaseValueItm = purchasedata_Itm["Price"].sum()

averageItemprice = totalPurchaseValueItm/purchasecnt

popularItm_dict = {"Purchase Count": purchasecnt, 
                   "Item Price": averageItemprice,
                   "Total Purchase Value": totalPurchaseValueItm
                    }
popularItem_df = pd.DataFrame(popularItm_dict)
popularItm_df = popularItem_df.sort_values("Purchase Count", ascending =  False)
popularItm_df["Item Price"] = popularItem_df["Item Price"].map("${:.2f}".format)
popularItm_df["Total Purchase Value"] = popularItem_df["Total Purchase Value"].map("${:.2f}".format)

popularItm_df.head()
```

```python
#Most Profitable Items
#Sort the above table by total purchase value in descending order
profitableItem_df = popularItem_df.sort_values(["Total Purchase Value"], ascending = False)
profitableItem_df["Item Price"] = profitableItem_df["Item Price"].map("${:.2f}".format)
profitableItem_df["Total Purchase Value"] = profitableItem_df["Total Purchase Value"].map("${:.2f}".format)

profitableItem_df.head()
```
